### EDA


```python
# droping unneccessary columns 
data.drop(columns=['Year_Birth','ID','Z_CostContact', 'Z_Revenue', 'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5'],inplace=True)
data
```

```python
data.hist(figsize=(20, 15), bins=30)
plt.suptitle("Distributions of Numerical Features")
plt.tight_layout()
plt.show()
```

```python
# outliers detection 
plt.figure(figsize=(16,16))
for i,col in enumerate(data,1):
    plt.subplot(5,6,i)
    sns.boxplot(data[col],color="red")
    plt.title(f"boxPlot of {col}")
    plt.grid()
plt.tight_layout()
plt.show()
```

```python
numeric_cols = data.select_dtypes(include='number').columns
numeric_cols
```

```python
numeric_cols = data.select_dtypes(include='number').columns
data_cleaned = data.copy()

for col in numeric_cols:
    Q1 = data_cleaned[col].quantile(0.25)
    Q3 = data_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    median_val = data_cleaned[col].median()
    
    # Replace outliers with median
    data_cleaned.loc[(data_cleaned[col] < lower_bound) | (data_cleaned[col] > upper_bound), col] = median_val
```

```python
data_cleaned.shape
```

```python

plt.figure(figsize=(12, 5))
sns.boxplot(data=data_cleaned[numeric_cols])
plt.title("After Outlier Removal (replacing outliers with median)")
plt.xticks(rotation=45)
plt.show()
```

```python

corr_matrix = data_cleaned.select_dtypes(include='number').corr()

plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, linewidths=.5)
plt.title('Correlation Heatmap of Numerical Features', fontsize=16)
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data_cleaned['Education'] = le.fit_transform(data_cleaned['Education'])

data_cleaned['Marital_Status'] = le.fit_transform(data_cleaned['Marital_Status'])
```

```python
data_cleaned
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
features_to_scale = ['Income', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth']
data_cleaned[features_to_scale] = scaler.fit_transform(data_cleaned[features_to_scale])
data_cleaned
```

```python
from sklearn.decomposition import PCA
data_cleaned['Dt_Customer'] = pd.to_datetime(data_cleaned['Dt_Customer']).apply(lambda date: date.toordinal())

pca = PCA() 
pc_data = pd.DataFrame(pca.fit_transform(data_cleaned))
pc_data.head()
```

```python
pc_data.var()
```

```python
x_feature = pc_data.iloc[:,0:2]
x_feature.head()
```

### KMEANS CLUSTERING


```python
from sklearn.cluster import KMeans

inertias = []
for i in range(1, 11):
    kmeans_pca = KMeans(n_clusters=i, random_state=0)
    kmeans_pca.fit(pc_data.iloc[:, :2])
    inertias.append(kmeans_pca.inertia_)

```

```python
# Plot elbow curve
plt.figure(figsize=(6,4))
plt.plot(range(1,11), inertias, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.xticks(range(1,11))
plt.show()
```

```python
pc_data.columns = pc_data.columns.astype(str)

kmeans_final = KMeans(n_clusters=4, random_state=0)
cluster_labels = kmeans_final.fit_predict(x_feature)

pc_data['Cluster'] = cluster_labels
```

```python

plt.figure(figsize=(8, 6))
sns.scatterplot(data=pc_data, x=pc_data.columns[0], y=pc_data.columns[1], 
                hue='Cluster', palette='tab10', s=60)
plt.title('Clusters on PCA components')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

score = silhouette_score(x_feature, cluster_labels)
print(f"Silhouette Score (2 PCs): {score:.4f}")
```

### hierarchical clustering 


```python
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import linkage, dendrogram
linked_ward = linkage(x_feature, method='ward')

plt.figure(figsize=(10, 5))
dendrogram(linked_ward)
plt.title("Dendrogram - Ward Linkage")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()

hc_ward = AgglomerativeClustering(n_clusters=4, linkage='ward')
pc_data['Ward_Cluster'] = hc_ward.fit_predict(x_feature)

plt.figure(figsize=(8, 5))
sns.scatterplot(x=pc_data.iloc[:, 0], y=pc_data.iloc[:, 1], hue='Ward_Cluster', data=pc_data, palette='Set1')
plt.title("Hierarchical Clustering (Ward)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.show()

score_ward = silhouette_score(x_feature, pc_data['Ward_Cluster'])
print(f"Silhouette Score (Ward): {score_ward:.4f}")
```

```python
linked_avg = linkage(x_feature, method='average')

plt.figure(figsize=(10, 5))
dendrogram(linked_avg)
plt.title("Dendrogram - Average Linkage")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()

hc_avg = AgglomerativeClustering(n_clusters=4, linkage='average')
pc_data['Avg_Cluster'] = hc_avg.fit_predict(x_feature)

plt.figure(figsize=(8, 5))
sns.scatterplot(x=pc_data.iloc[:, 0], y=pc_data.iloc[:, 1], hue='Avg_Cluster', data=pc_data, palette='Set2')
plt.title("Hierarchical Clustering (Average)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.show()

score_avg = silhouette_score(x_feature, pc_data['Avg_Cluster'])
print(f"Silhouette Score (Average): {score_avg:.4f}")
```

```python
linked_complete = linkage(x_feature, method='complete')

plt.figure(figsize=(10, 5))
dendrogram(linked_complete)
plt.title("Dendrogram - Complete Linkage")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()

hc_complete = AgglomerativeClustering(n_clusters=4, linkage='complete')
pc_data['Complete_Cluster'] = hc_complete.fit_predict(x_feature)

plt.figure(figsize=(8, 5))
sns.scatterplot(x=pc_data.iloc[:, 0], y=pc_data.iloc[:, 1], hue='Complete_Cluster', data=pc_data, palette='Set3')
plt.title("Hierarchical Clustering (Complete)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.show()

score_complete = silhouette_score(x_feature, pc_data['Complete_Cluster'])
print(f"Silhouette Score (Complete): {score_complete:.4f}")
```

```python
linked_single = linkage(x_feature, method='single')

plt.figure(figsize=(10, 5))
dendrogram(linked_single)
plt.title("Dendrogram - Single Linkage")
plt.xlabel("Samples")
plt.ylabel("Distance")
plt.show()

hc_single = AgglomerativeClustering(n_clusters=4, linkage='single')
pc_data['Single_Cluster'] = hc_single.fit_predict(x_feature)

plt.figure(figsize=(8, 5))
sns.scatterplot(x=pc_data.iloc[:, 0], y=pc_data.iloc[:, 1], hue='Single_Cluster', data=pc_data, palette='Dark2')
plt.title("Hierarchical Clustering (Single)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.grid(True)
plt.show()

score_single = silhouette_score(x_feature, pc_data['Single_Cluster'])
print(f"Silhouette Score (Single): {score_single:.4f}")
```

### DB SCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
from sklearn.neighbors import NearestNeighbors
import matplotlib.pyplot as plt
import numpy as np

# Step 1: K-distance graph to find suitable eps
neighbors = NearestNeighbors(n_neighbors=5)
neighbors_fit = neighbors.fit(data_cleaned)
distances, indices = neighbors_fit.kneighbors(data_cleaned)

distances = np.sort(distances[:, 4])  # 5th neighbor
plt.figure(figsize=(8, 4))
plt.plot(distances)
plt.title('K-Distance Graph')
plt.xlabel('Data Points sorted by distance')
plt.ylabel('5th Nearest Neighbor Distance')
plt.grid(True)
plt.show()
```

```python
dbscan = DBSCAN(eps=2.5, min_samples=5)
db_labels = dbscan.fit_predict(data_cleaned[features_to_scale])
data_cleaned['DBSCAN_Cluster'] = db_labels
```

```python
plt.scatter(data_cleaned[features_to_scale[0]], data_cleaned[features_to_scale[1]], c=data_cleaned['DBSCAN_Cluster'])
plt.title('DBSCAN Clusters')
plt.xlabel(features_to_scale[0])
plt.ylabel(features_to_scale[1])
plt.show()
```

```python
filtered_labels = data_cleaned.loc[data_cleaned['DBSCAN_Cluster'] != -1, 'DBSCAN_Cluster']
filtered_data = data_cleaned.loc[data_cleaned['DBSCAN_Cluster'] != -1].drop(columns=['DBSCAN_Cluster'])
num_clusters = len(set(filtered_labels))

if num_clusters > 1:
    score = silhouette_score(filtered_data, filtered_labels)
    print(f'Silhouette Score (DBSCAN): {score:.4f}')
else:
    print(f'Cannot calculate silhouette score: only {num_clusters} cluster(s) found after removing noise.')
```
